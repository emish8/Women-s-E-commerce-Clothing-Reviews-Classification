```python
from imblearn.under_sampling import RandomUnderSampler

def undersample(df, target  ):

    X = df.drop(target, axis=1)
    y = df[target]

    # Define the undersampler
    undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)

    # Apply undersampling to balance the classes
    X_resampled, y_resampled = undersampler.fit_resample(X, y)

    # Create a new DataFrame with the resampled data
    resampled_df = pd.concat([X_resampled, y_resampled], axis=1)
    
    return resampled_df
```

```python
all_data_resampled = undersample(all_data, 'positive')
only_text_resampled = undersample(only_text, 'positive')
original_data_resampled = undersample(original_data, 'positive')
```

```python
print(sum(all_data_resampled['positive'])/len(all_data_resampled['positive']))
print(sum(only_text_resampled['positive'])/len(only_text_resampled['positive']))
print(sum(original_data_resampled['positive'])/len(original_data_resampled['positive']))
```

```python
only_text_resampled.to_csv('only_text_resampled.csv', index=False)
all_data_resampled.to_csv('all_data_resampled.csv', index = False)
original_data_resampled.to_csv('original_data_resampled.csv', index = False)
```
